#1. Análisis de facturación



##1.1 Datos en cifras - Ingresos totales de las tiendas

```python
"""
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna
Precio de cada conjunto de datos de la tienda para estimar los ingresos.
"""

total_precio_tienda1 = tienda['Precio'].sum()
print(f"Total de precios para tienda 1: {total_precio_tienda1}")

total_precio_tienda2 = tienda2['Precio'].sum()
print(f"Total de precios para tienda 2: {total_precio_tienda2}")

total_precio_tienda3 = tienda3['Precio'].sum()
print(f"Total de precios para tienda 3: {total_precio_tienda3}")

total_precio_tienda4 = tienda4['Precio'].sum()
print(f"Total de precios para tienda 4: {total_precio_tienda4}")
```

##1.2 Graficos - Barras y Pastel - Tienda 4 menores ingresos que las demas

```python
import matplotlib.pyplot as plt
import seaborn as sns

totales_tiendas = {
    'Tienda 1': total_precio_tienda1,
    'Tienda 2': total_precio_tienda2,
    'Tienda 3': total_precio_tienda3,
    'Tienda 4': total_precio_tienda4
}

# Convertir a DataFrame para facilitar la visualización
df_totales = pd.DataFrame(list(totales_tiendas.items()), columns=['Tienda', 'Total Precio'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Total Precio', data=df_totales, palette='Paired')
plt.suptitle('Comparación de ingresos Totales por Tienda')
plt.title('La tienda 4 la que menos ingresos genera', fontsize=10, y=0.94, color='red')
plt.xlabel('Tienda')
plt.ylabel('Total de Ingresos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ticklabel_format(style='plain', axis='y') # Evita la notación científica en el eje Y
plt.ylim(600000000, 1200000000)
plt.savefig('comparacion_ingresos_Gbarra.png', dpi=300,bbox_inches='tight')
plt.show()
```

```python
df_totales = pd.DataFrame(list(totales_tiendas.items()), columns=['Tienda', 'Total Precio'])

paleta = ['#63E6C5', '#63B1E6', '#63E698', '#6388E6']
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    df_totales['Total Precio'],
    labels=df_totales['Tienda'],
    autopct='%1.1f%%',     # para mostrar porcentaje
    startangle=90,
    colors=paleta
)

for at in autotexts:
    at.set_color('white')
    at.set_fontweight('bold')

plt.suptitle('Proporción de Ingresos Totales por Tienda')
plt.title('La tienda 4 es la que menos ingresos genera', fontsize=10, color='red')
plt.tight_layout()
plt.savefig('comparacion_ingresos_Gpastel.png', dpi=300,bbox_inches='tight')
plt.show()
```

# 2. Ventas por categoría

## 2.1 Datos en cifras - Categorías de productos más y menos vendidas.

```python
"""
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría
y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.
"""

ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index()
print("Ventas por Categoría - Tienda 1:")
print(ventas_categoria_tienda1)

ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index()
print("\nVentas por Categoría - Tienda 2:")
print(ventas_categoria_tienda2)

ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index()
print("\nVentas por Categoría - Tienda 3:")
print(ventas_categoria_tienda3)

ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index()
print("\nVentas por Categoría - Tienda 4:")
print(ventas_categoria_tienda4)
```

```python
"""
Ordenados de mayor a menor
"""

tienda1_categorias = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Ventas por Categoría - Tienda 1:')
print(tienda1_categorias)

tienda2_categorias = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Ventas por Categoría - Tienda 1:')
print(tienda2_categorias)

tienda3_categorias = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Ventas por Categoría - Tienda 1:')
print(tienda3_categorias)

tienda4_categorias = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio',ascending=False)
print('Ventas por Categoría - Tienda 1:')
print(tienda4_categorias)

```

## 2.2 Graficos - Barras apiladas y barras de cada tienda

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Renombrar las columnas 'Precio' a los nombres de las tiendas para facilitar la combinación
t1 = ventas_categoria_tienda1.rename(columns={'Precio': 'Tienda 1'})
t2 = ventas_categoria_tienda2.rename(columns={'Precio': 'Tienda 2'})
t3 = ventas_categoria_tienda3.rename(columns={'Precio': 'Tienda 3'})
t4 = ventas_categoria_tienda4.rename(columns={'Precio': 'Tienda 4'})
# Unir por categoria cada tienda
df_totales_categorias = t1.merge(t2, on='Categoría del Producto') \
                          .merge(t3, on='Categoría del Producto') \
                          .merge(t4, on='Categoría del Producto')
# Establecer 'Categoría del Producto' como índice
df_totales_categorias = df_totales_categorias.set_index('Categoría del Producto')
df_tiendas = df_totales_categorias.T    # <- ESTA es la vuelta importante

# 5. Calculamos proporción por tienda (por fila ahora)
df_prop_tiendas = df_tiendas.div(df_tiendas.sum(axis=1), axis=0)
orden_categorias = df_prop_tiendas.sum(axis=0).sort_values(ascending=False).index


# esto te da un Index con las categorías ordenadas de menor a mayor
# 2. reordenar el dataframe con ese orden de columnas
df_prop_tiendas_ordenado = df_prop_tiendas[orden_categorias]
# 3. graficar ya ordenado
ax = df_prop_tiendas_ordenado.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 10)
)

plt.title('Proporción de Ingresos por Tienda y Categoría (ordenado)')
plt.xlabel('Tiendas')
plt.ylabel('Proporción de Ingresos')
plt.xticks(rotation=0)
plt.legend(title='Categorías', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda
